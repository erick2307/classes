# Numpy : [np.concatenate](https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html) ; [np.stack](https://numpy.org/doc/stable/reference/generated/numpy.stack.html) ; [np.split](https://numpy.org/doc/stable/reference/generated/numpy.split.html) ; [np.reshape](https://numpy.org/doc/stable/reference/generated/numpy.reshape.html)

```python
import numpy as np

# 例えば、以下のような2つの行列AとBがあったとします。
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(f'A={A}\n\nB={B}\n')

# これらの行列を結合して、以下のような行列ABを作成します。
A_B = np.concatenate((A, B), axis=0)
AB = np.concatenate((A, B), axis=1)
_AB = np.concatenate((A, B), axis=None)

# この行列ABを出力してください。
print(f'A_B={A_B}\n')
print(f'AB={AB}\n')
print(f'_AB={_AB}')
```

```python
# 2つの行列AとBを次のように積み重ねる必要があるとする。
C = np.stack((A,B),axis=0)
D = np.stack((A,B),axis=1)
print(f'C={C}\n\nD={D}')
```

```python
print(f'A_B={A_B.shape}, C={C.shape}, D={D.shape}')
```

```python
# 配列を複数の部分配列に分割する
print(f'_AB={_AB}\n')
print(np.split(_AB, 4))
```

```python
# 配列または行列の形状を変更する
print(f'_AB={_AB}\n')
print(f'_AB(shape)={_AB.shape}\n')
ABr = _AB.reshape(4,2)
print(f'ABr={ABr}\n')
```

```python
# function to stack to matrices and return the transpose
def concatenate_and_transpose(A, B):
    return np.concatenate((A,B),axis=0).T
```
